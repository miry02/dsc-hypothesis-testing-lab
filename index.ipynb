Now, choose and execute an appropriate statistical test. Make sure you describe why you chose the test, the null and alternative hypotheses, and what the result of the test means.


### Why Choose the t-test for Independent Samples?

Independent Samples
Numeric Outcome
Normality Assumption

### Null and Alternative Hypotheses:

Null Hypothesis:The average number of unhealthy days (`PHYSHLTH`) is the same for owners and renters.
  - \( \mu_{\text{Own}} = \mu_{\text{Rent}} \)

Alternative Hypothesis: The average number of unhealthy days (`PHYSHLTH`) differs between owners and renters.
  - \( \mu_{\text{Own}} \neq \mu_{\text{Rent}} \)

```python
# Your code here (create additional cells as needed)


### Execution of the t-test:


from scipy import stats

# Separate data for owners and renters
physhlth_own = df[df['RENTHOM1'] == 'Own']['PHYSHLTH']
physhlth_rent = df[df['RENTHOM1'] == 'Rent']['PHYSHLTH']

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(physhlth_own, physhlth_rent, equal_var=False)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

```

### Interpretation of Results:

- **T-statistic:** The t-statistic measures the difference between the means of owners and renters in units of standard error.
- **P-value:** The p-value indicates the probability of observing the data if the null hypothesis (that there is no difference in means) is true.

### Result Interpretation:

T-statistic: The t-statistic of -15.48 indicates a substantial difference in the average number of unhealthy days (PHYSHLTH) between homeowners and renters. A negative t-statistic suggests that, on average, renters report more unhealthy days than homeowners.

P-value: The extremely small p-value indicates strong evidence against the null hypothesis. 

## 4. Describe the Relationship between Chronic Sickness and Nicotine Use

Once again, this will require some preparation before we can run the statistical test. Create a new column `NICOTINE_USE` with 1 representing someone who uses or has used nicotine in some form, and 0 representing someone who hasn't.

We define nicotine use as:

* Answered `Yes` to the `SMOKE100` question (*Have you smoked at least 100 cigarettes in your entire life?*, page 43), OR
* Answered `Every day` or `Some days` to the `USENOW3` question (*Do you currently use chewing tobacco, snuff, or snus every day, some days, or not at all?*, page 46), OR
* Answered `Yes` to the `ECIGARET` question (*Have you ever used an e-cigarette or other electronic vaping product, even just one time, in your entire life?*, page 46)

If a record matches one or more of the above criteria, `NICOTINE_USE` should be 1. Otherwise, `NICOTINE_USE` should be 0. Go ahead and keep all of the "Don't know" or "Refused" answers as 0.

```python
# Your code here
# Replace missing or refused answers with 0 (not used nicotine)
df['SMOKE100'] = df['SMOKE100'].replace([7.0, 9.0], 0)
df['USENOW3'] = df['USENOW3'].replace([7.0, 9.0], 0)
df['ECIGARET'] = df['ECIGARET'].replace([7.0, 9.0], 0)

# Define function to determine nicotine use
def calculate_nicotine_use(row):
    if row['SMOKE100'] == 1 or row['USENOW3'] in [1, 2] or row['ECIGARET'] == 1:
        return 1
    else:
        return 0

# Create NICOTINE_USE column based on function
df['NICOTINE_USE'] = df.apply(calculate_nicotine_use, axis=1)
```

This time, let's treat health status as a categorical variable. We'll say that a "chronically sick" person is a person who reports that their physical health was not good for 15 or more out of the past 30 days. (This is a simplification but it will work for this analysis.)

In the cell below, create a new column of `df` called `CHRONIC`, which is 0 for records where `PHYSHLTH` is less than 15, and 1 for records where `PHYSHLTH` is 15 or more.

```python
# Your code here
# Define a function to categorize chronic sickness
def categorize_chronic(row):
    if row['PHYSHLTH'] >= 15:
        return 1
    else:
        return 0

# Create the CHRONIC column based on the function
df['CHRONIC'] = df.apply(categorize_chronic, axis=1)
```

Now we can view the crosstabs for these two categorical variables, as well as display their distributions:

```python
# Run this cell without changes
contingency_table = pd.crosstab(index=df["CHRONIC"], columns=df["NICOTINE_USE"])
contingency_table
```

```python
# Run this cell without changes

no_nicotine_use = df.loc[df["NICOTINE_USE"] == 0, "CHRONIC"]
nicotine_use = df.loc[df["NICOTINE_USE"] == 1, "CHRONIC"]

fig, ax = plt.subplots()

ax.hist(
    x=[no_nicotine_use, nicotine_use],
    label=["No Nicotine Use", "Nicotine Use"],
    bins=[0,1,2],
    align="left"
)

ax.set_ylabel("Count")
ax.set_xticks([0,1])
ax.set_xticklabels(["Not Chronically Sick", "Chronically Sick"])
ax.set_title("Distribution of Chronic Sickness by Nicotine Use")

ax.legend();
```

Once again, it appears that there is a difference in health outcomes between these categories. In the cell below, select an appropriate statistical test, describe the null and alternative hypotheses, execute the test, and interpret the result.

```python
# Your code here (create additional cells as needed)
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['NICOTINE_USE'], df['CHRONIC'])

# Perform chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p_value}")
```

Null Hypothesis (H₀): There is no association between nicotine use and chronic sickness.

The proportions of chronically sick individuals (CHRONIC == 1) are the same for both nicotine users and non-users.
Alternative Hypothesis (H₁): There is an association between nicotine use and chronic sickness.

The proportions of chronically sick individuals (CHRONIC == 1) differ between nicotine users and non-users.

The results of the chi-square test of independence are as follows:

- **Chi-square statistic:** 228.23
- **P-value:** \( 1.45 \times 10^{-51} \)

### Interpretation:

1. **Chi-square Statistic:** The chi-square statistic measures the magnitude of the association between nicotine use (`NICOTINE_USE`) and chronic sickness (`CHRONIC`). A higher chi-square value indicates a stronger association.

2. **P-value:** The extremely small p-value (\( 1.45 \times 10^{-51} \)) indicates strong evidence against the null hypothesis. This suggests that there is a statistically significant association between nicotine use and chronic sickness.

### Conclusion:

Based on these results, we reject the null hypothesis. This means there is strong evidence to support that there is an association between nicotine use and chronic sickness. Specifically, individuals who use nicotine are more likely to be chronically sick compared to those who do not use nicotine.
